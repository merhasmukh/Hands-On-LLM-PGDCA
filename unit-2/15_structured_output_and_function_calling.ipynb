## Part 1: Parsing JSON Mode Responses Safely
When requesting JSON from a model, it can occasionally output stray markdown characters or invalid JSON structures. We must handle parsing errors gracefully.


````python
import json

def parse_json_output(response_text):
    try:
        # Remove potential markdown fences
        cleaned = response_text.replace('```json', '').replace('```', '').strip()
        data = json.loads(cleaned)
        return data, 'Success'
    except json.JSONDecodeError as e:
        return None, f"Failed to parse JSON: {e}"

bad_json = '```json { "model": "gpt-4", "temperature": 0.5 } ```'
parsed_data, status = parse_json_output(bad_json)
print('Status:', status)
print('Parsed Data:', parsed_data)
````

## Part 2: OpenAI Function Calling Concepts
Function calling allows you to describe custom tools to an LLM. The model determines which tool to call and outputs the required arguments in JSON format.

## Hands-On Exercise
**Task**: Below is a raw invoice text. Create a JSON schema representation representing the structures we want to extract from it (vendor, invoice_date, total_amount).
Write a function `extract_invoice_fields(invoice_text)` that uses python dict matching or mock parsers to return a clean structured dictionary.


````python
invoice_raw = """
INVOICE #99102
Date: 2026-05-15
Vendor: Acme AI Corp
Line items:
 - Compute resources: $450.00
Total Amount Due: $450.00
"""

# TODO: Extract parameters from invoice_raw
def extract_invoice_fields(text):
    lines = text.strip().split('\n')
    vendor = 'Unknown'
    date = 'Unknown'
    amount = 'Unknown'
    
    for line in lines:
        if 'Vendor:' in line:
            vendor = line.split('Vendor:')[1].strip()
        elif 'Date:' in line:
            date = line.split('Date:')[1].strip()
        elif 'Total Amount Due:' in line:
            amount = line.split('Due:')[1].strip()
            
    return {
        'vendor': vendor,
        'invoice_date': date,
        'total_amount': amount
    }

print(extract_invoice_fields(invoice_raw))
````
