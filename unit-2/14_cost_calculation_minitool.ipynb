# Practical P14: Cost Calculation Mini-Tool
**Learning Outcome**: Reusable cost monitoring tool to add to any project.

## Part 1: Pricing Configuration Matrices
Let's define a pricing matrix containing input/output cost metrics for different models.


```python
PRICING_MATRIX = {
    'gpt-4o': {'input_1k': 0.005, 'output_1k': 0.015},
    'claude-3-5-sonnet': {'input_1k': 0.003, 'output_1k': 0.015},
    'gemini-1.5-pro': {'input_1k': 0.007, 'output_1k': 0.021}
}
```

## Part 2: Cost Estimations
Let's write a tracker to monitor API expenditures.


```python
class CostTracker:
    def __init__(self):
        self.total_cost = 0.0
        self.total_tokens = 0
        
    def record_call(self, model, prompt_tokens, completion_tokens):
        prices = PRICING_MATRIX.get(model, {'input_1k': 0.0, 'output_1k': 0.0})
        cost = (prompt_tokens / 1000.0) * prices['input_1k'] + (completion_tokens / 1000.0) * prices['output_1k']
        self.total_cost += cost
        self.total_tokens += (prompt_tokens + completion_tokens)
        return cost
```

## Hands-On Exercise
**Task**: Write a benchmark script that takes a fixed input token length (e.g. 500 prompt tokens) and completion length (e.g. 200 completion tokens) and outputs a pricing comparison summary table across the 3 models listed in the `PRICING_MATRIX`.


```python
# TODO: Output pricing comparison table
prompt_toks = 500
comp_toks = 200

print('='*55)
print('        MODEL COST COMPARISON REPORT')
print('='*55)
for model, rates in PRICING_MATRIX.items():
    tracker = CostTracker()
    cost = tracker.record_call(model, prompt_toks, comp_toks)
    print(f"Model: {model:<20} | Cost: ${cost:.6f} | CPM: ${(cost*1000):.4f}")
print('='*55)
```
