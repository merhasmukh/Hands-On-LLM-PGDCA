## Part 2: Defining batching operations
Processing 10,000 items in a single concurrent gather can overwhelm network rate limits.
Instead, we process items in smaller groups (batches) of size `N`.


```python
async def process_mock_batch(reviews_batch):
    # Simulates query
    results = []
    for r in reviews_batch:
        sentiment = 'POSITIVE' if 'beautiful' in r or 'ergonomic' in r or 'Exceptional' in r else 'NEGATIVE'
        results.append(sentiment)
    return results
```

## Hands-On Exercise
**Task**: Write an async pipeline that processes the rows of `data/product_reviews.csv` in batches of exactly 5. Classify the review text as `'POSITIVE'` or `'NEGATIVE'`, append the predictions as a new column, and save the output to `data/product_reviews_classified.csv`.


```python
# TODO: Complete batch processing loops
async def run_batch_classification():
    df_reviews = pd.read_csv('data/product_reviews.csv')
    batch_size = 5
    predictions = []
    
    for i in range(0, len(df_reviews), batch_size):
        batch_text = df_reviews['text'].iloc[i:i+batch_size].tolist()
        results = await process_mock_batch(batch_text)
        predictions.extend(results)
        print(f'Processed batch {i // batch_size + 1}')
        
    df_reviews['predicted_sentiment'] = predictions
    df_reviews.to_csv('data/product_reviews_classified.csv', index=False)
    print('Finished bulk classification.')
    
await run_batch_classification()
pd.read_csv('data/product_reviews_classified.csv').head(5)
```
