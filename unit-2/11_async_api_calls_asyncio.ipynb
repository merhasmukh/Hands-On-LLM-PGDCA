# Practical P11: Async API Calls with asyncio
**Learning Outcome**: Make concurrent API calls using `asyncio` to speed up batch tasks.

## Part 1: What is Asynchronous Programming?
Normally, scripts run synchronously (one line at a time). When waiting for network API completions, the CPU sits idle.
Asynchronous programming allows your script to work on other queries while waiting for a response, processing multiple prompts simultaneously.


```python
import asyncio
import time

async def mock_async_api_call(prompt, delay=1.0):
    # Simulated non-blocking network wait using asyncio.sleep
    await asyncio.sleep(delay)
    return f"Processed: '{prompt}'"

# Test a single async call
async def main():
    print('Starting async query...')
    res = await mock_async_api_call('Test prompt')
    print('Result:', res)

# Since jupyter runs its own event loop, we execute using await directly in the cells
await main()
```

## Part 2: Concurrency with `asyncio.gather`
To send multiple requests concurrently, we pack tasks into a list and execute them together.


```python
async def batch_concurrent_queries():
    prompts = ['Prompt A', 'Prompt B', 'Prompt C', 'Prompt D', 'Prompt E']
    tasks = [mock_async_api_call(p, delay=1.5) for p in prompts]
    
    t_start = time.time()
    results = await asyncio.gather(*tasks)
    duration = time.time() - t_start
    
    print(f'Successfully completed {len(results)} queries in {duration:.2f} seconds!')
    print('Results:', results)

await batch_concurrent_queries()
```
