## Part 1: How Parameters Control Output Generation
LLM APIs use probability distributions to select the next token. Parameters adjust these distributions:
- **Temperature**: High temperature = more random/creative; low temperature = more deterministic.
- **Top-p**: Nucleus sampling cutoff (selects from top percentile cumulative probabilities).
- **Max Tokens**: Restricts the maximum tokens the model can generate.
- **Stop Sequences**: String markers that tell the model to halt generation.


```python
def query_mock_params(prompt, temp=0.7, top_p=0.9, max_tokens=50, stop=None):
    # Mock simulation of parameters effects
    if temp == 0:
        reply = 'Deterministic output. 2+2 is 4. Coding is logic.'    
    elif temp > 1.2:
        reply = 'Chaotic dreamscapes! Fluffy codes run through pythonic jungles of random words.'    
    else:
        reply = f"Balanced output answering prompt: '{prompt}' with temp={temp}."
        
    # Simulate max_tokens truncation
    words = reply.split()
    if len(words) > max_tokens:
        reply = ' '.join(words[:max_tokens]) + ' [TRUNCATED]'
        
    # Simulate stop sequences
    if stop:
        for s in stop:
            if s in reply:
                reply = reply.split(s)[0]
                
    return reply

print('Temp 0 (Deterministic):', query_mock_params('Tell me a programming joke', temp=0))
print('Temp 1.5 (Creative/Chaotic):', query_mock_params('Tell me a programming joke', temp=1.5))
```
