## Part 2: Loading Models via Transformers Pipelines
HuggingFace provides simple wrappers to download and query weights in minutes.


```python
# Mock pipeline setup
def mock_hf_pipeline(task, model_name):
    print(f"Downloading model files for '{model_name}' from HuggingFace Hub...")
    print(f"Setting up pipeline configuration for task: {task}")
    return lambda prompt: f"[Model {model_name} generated]: Output responding to prompt: '{prompt}'"

generator = mock_hf_pipeline('text-generation', 'distilgpt2')
print(generator('AI is'))
```
