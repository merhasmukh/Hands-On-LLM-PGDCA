# Practical P23: No-code: Build a Chatbot in Flowise
**Learning Outcome**: Experience no-code AI development vs code-first.

## Part 1: visual Chatflows
Low-code/No-code tools like Flowise allow you to build RAG pipelines visually by connecting nodes (Document Loader -> Vector DB -> LLM Chain -> User UI).

### Commands to install and run Flowise locally:
```bash
npx flowise start
```
Then open `http://localhost:3000` in your browser to build workflows.
