## Part 1: Multimodal Vision Tokenization Flow
Vision LLMs combine visual encoders (like CLIP) with text decoders, translating images into sequential arrays of vision tokens.

```mermaid
graph TD
  ImageInput[Image: Receipt Photo] --> VisionEncoder[Visual Transformer / ViT]
  TextInput[Text prompt: extract total] --> TextTokenizer[Text Tokenizer]
  VisionEncoder --> VisionTokens[Visual Tokens: [v1, v2...]]
  TextTokenizer --> TextTokens[Text Tokens: [t1, t2...]]
  VisionTokens --> JointContext[Concatenated Input Context]
  TextTokens --> JointContext --> Dec[Autoregressive Decoder LLM] --> Output[JSON Invoice Details]
```
