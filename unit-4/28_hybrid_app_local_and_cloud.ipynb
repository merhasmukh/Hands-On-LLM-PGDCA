**Learning Outcome**: Build a production-relevant hybrid AI system that routes queries intelligently.

## Part 1: Hybrid Routing Module
Let's write a python module `hybrid_router.py` that decides whether to resolve prompts locally (for privacy/cost) or query cloud resources (for complex logic tasks).


```python
%%writefile hybrid_router.py
# hybrid_router.py - Smart Query Router
class HybridAIRouter:
    def __init__(self, threshold_words=20):
        self.threshold = threshold_words
        self.local_count = 0
        self.cloud_count = 0
        
    def route_query(self, prompt):
        word_length = len(prompt.split())
        # Simple routing rule: long/complex prompts go cloud, simple requests go local
        if word_length > self.threshold or 'summarize' in prompt.lower() or 'analyze' in prompt.lower():
            self.cloud_count += 1
            reply = f"[Resolved by Cloud Model (GPT-4o)]: Processed complex request '{prompt}'"
            route = 'CLOUD'
        else:
            self.local_count += 1
            reply = f"[Resolved by Local Model (Llama 3.2)]: Processed simple request '{prompt}'"
            route = 'LOCAL'
            
        return reply, route
```

## Part 2: Verifying the Router logic
Let's load and run test prompts to verify routing rules.


```python
import hybrid_router
from hybrid_router import HybridAIRouter

router = HybridAIRouter(threshold_words=10)
prompts = [
    'Hello assistant.', # short -> local
    'Explain deep learning in detail and output the mathematical equations.', # long -> cloud
    'Summarize this sentence.', # keyword summary -> cloud
]

for p in prompts:
    ans, route = router.route_query(p)
    print(f"Prompt: '{p}' -> Routed to: {route}")
```
