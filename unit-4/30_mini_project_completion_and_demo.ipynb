# Practical P30: Mini Project: Completion & Final Presentation
**Learning Outcome**: End-to-end LLM application presented as semester capstone.

## Capstone Checklist
Before launching live group presentations, review details against this scorecard:
1. **Core Functionality (40%)**: Does the pipeline load data, query models, maintain memory or parse JSON structures correctly?
2. **Code Quality (30%)**: Are variables structured inside clean classes? Are error states handled gracefully? Is logging included?
3. **Presentation & Pitch (30%)**: Can you explain system designs and highlight one integration challenge resolved by the team?
