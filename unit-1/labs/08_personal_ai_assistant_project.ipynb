```python
import os
from dotenv import load_dotenv

class InteractiveConsoleBot:
    def __init__(self, system_persona):
        self.system_prompt = system_persona
        self.history = [{'role': 'system', 'content': self.system_prompt}]
        self.total_tokens_used = 0
        
    def query_response(self, user_text):
        self.history.append({'role': 'user', 'content': user_text})
        
        # Mock Response matching OpenAI usage format
        reply = f"[Response as AI Persona] Echoing message: '{user_text}'"
        tokens = len(user_text.split()) + 20
        self.total_tokens_used += tokens
        
        self.history.append({'role': 'assistant', 'content': reply})
        return reply, tokens
```
