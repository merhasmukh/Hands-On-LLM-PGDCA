## Part 1: Lists and Loop Operations
We represent sequential chat interactions as lists of dictionaries in Python. Let's see how to add, slice, and loop over messages.


````python
chat_messages = []

# Add a user prompt to chat history
chat_messages.append({'role': 'user', 'content': 'What is a neural network?'})
# Add model response
chat_messages.append({'role': 'assistant', 'content': 'A computing system inspired by biological brains.'})

# Loop through and print transcript formatted style
for msg in chat_messages:
    role_label = msg['role'].upper()
    print(f"{role_label}: {msg['content']}")
````

## Part 3: Python Functions and f-strings
Functions keep code reusable. Let's wrap prompts formatting inside helper functions.


````python
def format_completion_prompt(instruction, input_text):
    """
    Encapsulates prompt engineering variables into structured string templates.
    """
    return f"System: {instruction}\nUser: {input_text}\nOutput:"

p = format_completion_prompt('Translate to French', 'Good morning, how are you?')
print(p)
````

**Task**: Write a function `build_chat_message(role, content)` that returns a dictionary. Write a second function `add_message_and_format(history_list, role, content)` that appends the new dictionary to the history list and prints the entire history on screen in a readable conversation layout.


````python
# TODO: Complete both functions and run checks
def build_chat_message(role, content):
    return {'role': role, 'content': content}

def add_message_and_format(history_list, role, content):
    msg = build_chat_message(role, content)
    history_list.append(msg)
    for index, m in enumerate(history_list):
        print(f"[{index+1}] {m['role'].capitalize()}: {m['content']}")

history = []
add_message_and_format(history, 'system', 'You are an AI assistant.')
print('-'*40)
add_message_and_format(history, 'user', 'What is Python?')
````
