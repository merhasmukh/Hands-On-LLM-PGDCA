## Part 1: System Prompts and Personas
A system prompt configures the behavior, tone, style, and scope of the assistant before user inputs are processed.
Let's see a prompt constructor.


```python
def get_system_prompt(persona):
    if persona == 'chef':
        return 'You are a Michelin-star Italian chef. Answer culinary questions using cooking terms.'
    elif persona == 'coder':
        return 'You are an elite Python developer. Respond with code blocks and short explanations.'
    else:
        return 'You are a friendly general assistant.'

print('Chef Persona Prompt:\n', get_system_prompt('chef'))
```

## Part 2: Few-Shot Prompting
Few-shot prompting provides examples to guide the model's output formatting.


```python
few_shot_prompt = """
Instruction: Convert review to sentiment label.

Review: I absolute loved this tool!
Sentiment: POSITIVE

Review: It was quite slow and timed out.
Sentiment: NEGATIVE

Review: Decent performance, but overpriced.
Sentiment: NEUTRAL

Review: {user_review}
Sentiment:
"""

formatted = few_shot_prompt.format(user_review='This software crashed within 5 minutes.')
print(formatted)
```
