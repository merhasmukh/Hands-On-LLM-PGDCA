## Part 2: Tokenizing Text using string estimation
Let's write a tokenization estimator helper.


````python
def estimate_token_count(text):
    # Rule of thumb: 1 token is roughly 4 characters or 0.75 words
    word_tokens = int(len(text.split()) * 1.3)
    char_tokens = int(len(text) / 4)
    return max(word_tokens, char_tokens)

sentence = 'Deep learning models use transformers to analyze complex languages.'
print(f"Words: {len(sentence.split())} | Estimated Tokens: {estimate_token_count(sentence)}")
````

## Part 3: Retrieving Usage Stats from API Responses
Every model execution returns usage metrics in the JSON structure. Let's inspect them.


````python
mock_response = {
    'id': 'chatcmpl-99',
    'usage': {
        'prompt_tokens': 12,
        'completion_tokens': 24,
        'total_tokens': 36
    }
}

usage = mock_response['usage']
print('Prompt Tokens size:', usage['prompt_tokens'])
print('Completion response tokens:', usage['completion_tokens'])
print('Total request consumption:', usage['total_tokens'])
````

````python
# TODO: Complete print_token_billing_report
def print_token_billing_report(resp_dict, cost_per_1k=0.002):
    usage = resp_dict.get('usage', {})
    total = usage.get('total_tokens', 0)
    cost = (total / 1000.0) * cost_per_1k
    
    print('+' + '-'*35 + '+')
    print(f"| Prompt Tokens:      {usage.get('prompt_tokens'):<13} |")
    print(f"| Completion Tokens:  {usage.get('completion_tokens'):<13} |")
    print(f"| Total Tokens:       {total:<13} |")
    print(f"| Estimated Cost:     ${cost:<12.6f} |")
    print('+' + '-'*35 + '+')

print_token_billing_report(mock_response)
````
