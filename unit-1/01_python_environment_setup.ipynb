### Debugging Common Syntax Mistakes
Indentation is syntactically meaningful in Python. Incorrect spacings trigger an `IndentationError`.


```python
try:
    # Triggering an indentation syntax exception programmatically using eval
    eval("""\ndef test_indent():\n  a = 1\n   b = 2 # Wrong spacing""")
except IndentationError as err:
    print('Caught IndentationError: Always align code blocks inside functions!')
```
