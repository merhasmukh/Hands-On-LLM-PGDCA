# Practical P5: Exploring Multiple API Providers
**Learning Outcome**: Switch comfortably between API providers.

## Part 1: Schema Inconsistencies between SDKs
OpenAI, Anthropic, and Google Gemini expose different Python client objects and require different key formats. Let's see them side-by-side:
- **OpenAI**: uses `openai.OpenAI()` client and `choices[0].message.content` response parsing.
- **Anthropic**: uses `anthropic.Anthropic()` client and `content[0].text` response parsing.
- **Gemini**: uses `google.generativeai.GenerativeModel()` and `response.text` response parsing.


## Part 2: Building an Adapter Wrapper Class
To solve this, developers build adapter classes to expose a single unified interface.


```python
class UnifiedAIAdapter:
    def __init__(self, provider, api_key):
        self.provider = provider
        self.api_key = api_key
        
    def generate(self, prompt):
        # Simulate schema translations
        if self.provider == 'openai':
            return f"[OpenAI Response] Echo prompt: '{prompt}'"
        elif self.provider == 'anthropic':
            return f"[Anthropic Claude Response] Echo prompt: '{prompt}'"
        elif self.provider == 'gemini':
            return f"[Google Gemini Response] Echo prompt: '{prompt}'"
        else:
            raise ValueError('Unknown Provider')
```

```python
openai_client = UnifiedAIAdapter('openai', 'sk-mock')
gemini_client = UnifiedAIAdapter('gemini', 'sk-mock')

print(openai_client.generate('What is AI?'))
print(gemini_client.generate('What is AI?'))
```
